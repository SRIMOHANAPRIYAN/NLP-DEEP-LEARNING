```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
def senToken(sentence):
    # dataset=text
    sentences=[text.split() for text in sentence]
    # print(sentences)
    return sentences
```

```python
engSentences=senToken(english_sentence)
```

```python
engSentences
```

```python
tamSentences=senToken(tamil_sentence)
```

```python
tamSentences
```

```python
def ownModel(langSentence,modelname):
    model=Word2Vec(langSentence,min_count=1)
    print(model)
    model.save(modelname)
    X = model.wv.vectors
    pca = PCA(n_components=2)
    result = pca.fit_transform(X)
    plt.scatter(result[:, 0], result[:, 1])
    words = list(model.wv.index_to_key)
    for i, word in enumerate(words):
        plt.annotate(word, xy=(result[i, 0], result[i, 1]))
    plt.show()
```

```python
ownModel(engSentences,'engmodel.bin')
```
